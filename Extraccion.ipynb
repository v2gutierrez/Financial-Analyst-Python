```python
import yfinance as yf
import pandas as pd

# Configuramos Pandas para ver todas las columnas
pd.set_option('display.max_columns', None)

# Seleccionamos una empresa del Mercado Continuo Español, por ejemplo, Inditex (ITX.MC)
ticker_symbol = "ITX.MC"
empresa = yf.Ticker(ticker_symbol)

print(f"--- Información General de {ticker_symbol} ---")
# print(empresa.info) # Descomenta para ver muchísima info

print("\n--- Balance (Balance Sheet) ---")
balance = empresa.balance_sheet
display(balance) # 'display' funciona mejor que 'print' en Jupyter

print("\n--- Cuenta de Resultados (Financials) ---")
cuenta_resultados = empresa.financials
display(cuenta_resultados)

# Guarda tu notebook!
```
